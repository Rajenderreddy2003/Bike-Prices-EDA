```python
brands = ['RoyalEnfield', 'Honda', 'Bajaj', 'TVS', 'Yamaha', 'Hero', 'Kawasaki', 'KTM', 'Suzuki', 'Triumph', 'Ampere', 'Aprilia', 'Ather', 'Benelli', 'Benling', 'BGauss', 'BMW', 'Bounce', 'Brixton-Motorcycles', 'BSA', 'CFMoto', 'Ducati', 'Evolet', 'Ferrato', 'Gemopai', 'HarleyDavidson', 'Hop-Electric', 'Husqvarna', 'Indian', 'iVOOMi', 'Jawa', 'Joyebike', 'Keeway', 'Kinetic-Green', 'Lambretta', 'Lectrix', 'Matter', 'MotoGuzzi', 'MotoMorini', 'Numeros-Motors', 'Oben', 'Odysse', 'Okaya', 'Okinawa', 'OLA', 'PUREEV', 'QJMotor', 'Quantum-Energy', 'Revolt', 'River', 'Simple-Energy', 'Ultraviolette', 'Vespa', 'VIDA', 'Yezdi', 'Yo', 'Zontes']
upcoming_bikes = []
for i in brands:
    i = i.lower()
    url = f"https://www.bikewale.com/{i}-bikes/upcoming/"
    page = requests.get(url)
    soup = BeautifulSoup(page.text)
    table = soup.find("table")
    if table:
       for row in table.find_all("tr")[1:]:  # skip header row
            cols = row.find_all("td")
            if cols:
                name = cols[0].text.strip()
                upcoming_bikes.append(name)

    # Fallback: also check bullet-list sections
    if not upcoming_bikes:
        for li in soup.select("ul li"):
            txt = li.text.strip()
            if "Royal Enfield" in txt:
                upcoming_bikes.append(txt)
```

```python
brands = ['RoyalEnfield', 'Honda', 'Bajaj', 'TVS', 'Yamaha', 'Hero', 'Kawasaki', 'KTM', 'Suzuki', 'Triumph', 'Ampere', 'Aprilia', 'Ather', 'Benelli', 'Benling', 'BGauss', 'BMW', 'Bounce', 'Brixton-Motorcycles', 'BSA', 'CFMoto', 'Ducati', 'Evolet', 'Ferrato', 'Gemopai', 'HarleyDavidson', 'Hop-Electric', 'Husqvarna', 'Indian', 'iVOOMi', 'Jawa', 'Joyebike', 'Keeway', 'Kinetic-Green', 'Lambretta', 'Lectrix', 'Matter', 'MotoGuzzi', 'MotoMorini', 'Numeros-Motors', 'Oben', 'Odysse', 'Okaya', 'Okinawa', 'OLA', 'PUREEV', 'QJMotor', 'Quantum-Energy', 'Revolt', 'River', 'Simple-Energy', 'Ultraviolette', 'Vespa', 'VIDA', 'Yezdi', 'Yo', 'Zontes']
brand = []
model_name = []
cc = []
mileage = []
price = []
bhp = []
weight = []
ratings = []
Type = []
for i in brands:
    i = i.lower()
    url = f"https://www.bikewale.com/{i}-bikes/"
    page = requests.get(url)
    soup = BeautifulSoup(page.text)
    brand_found = soup.find_all("h3",class_ = "o-j4 o-jq o-hM o-c4")
    for i in brand_found:
        text = i.text
        if text in upcoming_bikes:
            continue
        if "Royal" in text:
            text = "Royal-Enfield"
        elif "Brixton" in text:
            text = "Brixton-Motorcycles"
        elif "Hop" in text:
            text = "Hop-Electric"
        elif "Joy e-bike" in text:
            text = "Joy-e-bike"
        elif "Kinetic Green" in text:
            text = "Kinetic-Green"
        elif "Moto Guzzi" in text:
            text = "Moto-Guzzi"
        elif "Moto Morini" in text:
            text = "Moto-Morini"
        elif "Numeros Motors" in text:
            text = "Numeros-Motors"
        elif "PURE EV" in text:
            text = "PURE-EV"
        elif "QJ Motor" in text:
            text = "QJ-Motor"
        elif "Quantum Energy" in text:
            text = "Quantum-Energy"
        elif "Simple Energy" in text:
            text = "Simple-Energy"
        else:
            text = text.split()[0]
        brand.append(text)
    
    models_found = soup.find_all("h3",class_ = "o-j4 o-jq o-hM o-c4")
    for i in models_found:
        if i.text not in upcoming_bikes:
            model_name.append(i.text)
    cc_found = soup.find_all("div",class_ = "o-o onOEBQ o-iT o-ei")
    for i in cc_found:
        j = re.findall(r"(\d+|\d+\.\d+)\scc",i.text)
        if j:
            cc.append(j[0])
        else:
            cc.append(0)
    mileage_found = soup.find_all("div",class_ = "o-o onOEBQ o-iT o-ei")
    for i in mileage_found:
        j = re.findall(r"(\d+(?:\.\d+)?)\s*(?=kmpl|km\b)",i.text)
        if j:
            mileage.append(j[0])
        else:
            mileage.append(np.nan)
    price_found = soup.find_all("span",class_ = "o-jJ o-jr o-j3 o-kJ")
    for i in price_found:
        price.append(i.text)
    bhp_found = soup.find_all("div",class_ = "o-o onOEBQ o-iT o-ei")
    for i in bhp_found:
        j = re.findall(r"(\d+|\d+\.\d+)\sbhp",i.text)
        if j:
            bhp.append(j[0])
        else:
            bhp.append(np.nan)
    weight_found = soup.find_all("div",class_ = "o-o onOEBQ o-iT o-ei")
    for i in weight_found:
        j = re.findall(r"(\d+|\d+\.\d+)\skg",i.text)
        if j:
            weight.append(j[0])
        else:
            weight.append(np.nan)
    #  ratings out of 5
    ratings_tags = soup.find_all("div", class_="o-C o-cF o-c9 o-bT o-cp o-f")
    for i in ratings_tags:
        rating_match = i.find("p", class_="o-iY o-cx o-jq o-d o-j o-jJ")
        text = i.text.strip()

        if text.lower().startswith("upcoming"):
            continue
        if rating_match:
            match = re.search(r"(\d(?:\.\d)?)/\d", rating_match.text)
            if match:
                ratings.append(match.group(1))
            else:
                ratings.append(np.nan)
        else:
            ratings.append(np.nan)
    # fuel_type
    for i in soup.find_all("div" , class_="o-C o-cF o-c9 o-bT o-cp o-f"):
        text = i.text.strip()

        # Skip if Upcoming
        if text.lower().startswith("upcoming"):
             continue
        # Check if EV or Petrol based on features
        if text.startswith("EV") or re.search(r'\d+\s?kmph.*?\d+\s?km.*?\d+\s?k[wW]', text):
            bikeType = "EV"
        elif text.startswith("CNG"):
            bikeType ="CNG"
        elif re.search(r'\d+\.?\d*\s?(cc|bhp)', text, re.IGNORECASE):
            bikeType = "Petrol"
        else:
            bikeType= np.nan
        Type.append(bikeType)
```

```python
data = {
    "Brand": brand,
    "Model_Name": model_name,
    "CC": cc,
    "BHP":bhp,
    "Mileage":mileage,
    "Price": price,
    "Type": Type,
    "Weight":weight,
    "Rating":ratings
}

df = pd.DataFrame(data)
df = df.drop(df[df['Type'] == 'CNG'].index)
df.to_csv("bikes_data.csv", index=False, encoding='utf-8-sig')
```
